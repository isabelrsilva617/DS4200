```python
import matplotlib.pyplot as plt
import numpy as np
 
x=range(1,41)
values=np.random.uniform(size=40)
 
x_sorted = sorted(x)
values_sorted = sorted(values)

# stem function
plt.stem(x, values)
plt.ylim(0, 1.2)
plt.show()

plt.stem(x_sorted, values_sorted)
plt.ylim(0, 1.2)
plt.show()
```

```python
# Plot the lollipops horizontally
x=range(1,41)
values=np.random.uniform(size=40)

x_sorted = sorted(x)
values_sorted = sorted(values)

plt.hlines(y=x_sorted, xmin=0, xmax = values_sorted, color='green')
plt.plot(values_sorted, x_sorted, '*', color="red")
plt.show()
```
